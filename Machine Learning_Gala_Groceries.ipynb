```python
def convert_to_datetime(data: pd.DataFrame = None, column: str = None):

  dummy = data.copy()
  dummy[column] = pd.to_datetime(dummy[column], format='%Y-%m-%d %H:%M:%S')
  return dummy
```

```python
sales_df = convert_to_datetime(sales_df, 'timestamp')
sales_df.info()
```

```python
stock_df = convert_to_datetime(stock_df, 'timestamp')
stock_df.info()
```

```python
temp_df = convert_to_datetime(temp_df, 'timestamp')
temp_df.info()
```

```python
from datetime import datetime

def convert_timestamp_to_hourly(data: pd.DataFrame = None, column: str = None):
  dummy = data.copy()
  new_ts = dummy[column].tolist()
  new_ts = [i.strftime('%Y-%m-%d %H:00:00') for i in new_ts]
  new_ts = [datetime.strptime(i, '%Y-%m-%d %H:00:00') for i in new_ts]
  dummy[column] = new_ts
  return dummy
```

```python
sales_df = convert_timestamp_to_hourly(sales_df, 'timestamp')
sales_df.head()
```

```python
stock_df = convert_timestamp_to_hourly(stock_df, 'timestamp')
stock_df.head()
```

```python
temp_df = convert_timestamp_to_hourly(temp_df, 'timestamp')
temp_df.head()
```

```python
sales_df1 = sales_df.groupby(['timestamp', 'product_id']).agg({'quantity': 'sum'}).reset_index()
sales_df1.head()
```

```python
stock_df1 = stock_df.groupby(['timestamp', 'product_id']).agg({'estimated_stock_pct': 'mean'}).reset_index()
stock_df1.head()
```

```python
temp_df1 = temp_df.groupby(['timestamp']).agg({'temperature': 'mean'}).reset_index()
temp_df1.head()
```

```python
dataframe = stock_df1.merge(sales_df1, on=['timestamp', 'product_id'], how='left')
dataframe.head()
```

```python
dataframe_1 = dataframe.merge(temp_df1, on='timestamp', how='left')
dataframe_1.head()
```

```python
dataframe_1['quantity'] = dataframe_1['quantity'].fillna(0)
dataframe_1.info()
```

```python
product_cat = sales_df[['product_id', 'category']]
product_cat = product_cat.drop_duplicates()
```
